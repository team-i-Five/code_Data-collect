## 상세 내용 가져오기 테스트

```python
detail_url = "http://www.playdb.co.kr/playdb/playdbDetail.asp?sReqPlayno=50653"

# synopsis_url = "https://www.playdb.co.kr/playdb/playdbDetail_Content.asp?TabKind=2&PlayNo=50653"
synopsis_url = "https://www.playdb.co.kr/playdb/playdbDetail_Content.asp?TabKind=2&PlayNo=192992"
```

```python
r1 = requests.get(detail_url)
r2 = requests.get(synopsis_url)
r1, r2
```

```python
soup1 = BeautifulSoup(r1.text,'html.parser')
soup1
```

```python
soup2 = BeautifulSoup(r2.text,'html.parser')
soup2
```

```python
# 작품설명
soup2.find_all("td",class_="news")[0].text
```

```python
# 줄거리
soup2.find_all("td",class_="news")[1]
```

```python
td_elements = soup2.find_all("td")
td_elements
```

```python
# 필요한 <td> 요소 선택
for td_element in td_elements:
    a_element = td_element.find("a", {"name": "TabSynopsis"})  # 해당하는 <a> 태그 찾기
    if a_element:
        text = td_element.find("td", class_="news").get_text()  # 원하는 <td> 안의 텍스트 가져오기
        print(text)
```
