## 상세 내용 가져오기 테스트

```python
detail_url = "http://www.playdb.co.kr/playdb/playdbDetail.asp?sReqPlayno=3635"

synopsis_url = "https://www.playdb.co.kr/playdb/playdbDetail_Content.asp?TabKind=2&PlayNo=50653"
```

```python
r1 = requests.get(detail_url)
r2 = requests.get(synopsis_url)
r1, r2
```

```python
soup1 = BeautifulSoup(r1.text,'html.parser')
soup1
```

```python
soup2 = BeautifulSoup(r2.text,'html.parser')
soup2
```

```python
# 작품설명
soup2.find_all("td",class_="news")[0].text
```

```python
# 줄거리
soup2.find_all("td",class_="news")[1]
```

```python
soup1.find("div", class_="pddetail").find("h2").find("img").get("src")
```

```python
try:
    poster = soup1.find("div", class_="pddetail").find("h2").find("img").get("src")
    genre = soup1.find("img", alt="세부장르").find_next("td").get_text(strip=True)
    date = soup1.find("img", alt="일시").find_next("td").get_text(strip=True)
    location = soup1.find("img", alt="장소").find_next("td").get_text(strip=True)
    cast = soup1.find("img", alt="출연").find_next("td").get_text(strip=True)
    age_rating = soup1.find("img", alt="관람등급").find_next("td").get_text(strip=True)
    running_time = soup1.find("img", alt="관람시간").find_next("td").get_text(strip=True)
except AttributeError as ae:
    # 에러 난 변수를 null 값으로 할당하고 싶은데
    print(ae)

print(poster)
print(genre)
print(date)
print(location)
print(cast)
print(age_rating)
print(running_time)
```
