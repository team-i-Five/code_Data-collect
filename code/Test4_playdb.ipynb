```python
r_list = requests.get(url)
```

```python
soup = BeautifulSoup(r_list.text, 'html.parser')
soup
```

```python
for i in range(1, 10):
    a_tags_with_onclick = soup.find_all('a', onclick=True)
    playno = a_tags_with_onclick[i]['onclick'].split("'")[1] if len(a_tags_with_onclick) > 1 else None
    
    play_url = f"http://www.playdb.co.kr/playdb/playdbDetail.asp?sReqPlayno={playno}"
    story_url = f"http://www.playdb.co.kr/playdb/playdbDetail_Content.asp?TabKind=2&PlayNo={playno}"
    
    print(play_url)
    print(story_url)
    print("==================================================================================")
    
```

```python
# 각 정보를 추출하여 변수에 저장
genre = soup.find("img", alt="세부장르").find_next("td").get_text(strip=True)
date = soup.find("img", alt="일시").find_next("td").get_text(strip=True)
location = soup.find("img", alt="장소").find_next("td").get_text(strip=True)
cast = soup.find("img", alt="출연").find_next("td").get_text(strip=True)
# age_rating = soup.find("img", alt="관람등급").find_next("td").get_text(strip=True)
running_time = soup.find("img", alt="관람시간").find_next("td").get_text(strip=True)

# 변수에 저장된 정보 출력
print("장르:", genre)
print("일시:", date)
print("장소:", location)
print("출연:", cast)
# print("관람등급:", age_rating)
print("관람시간:", running_time)
```
